```python
#get columns to plot
columns=iris.columns.drop(['species'])
#create the dataset
x_data=range(0,iris.shape[0])
#create a figure and axis
fig,ax=plt.subplots()

#plot each column
for column in columns:
  ax.plot(x_data,iris[column],label=column)
ax.set_title('Iris Dataset')
ax.legend()
```

```python
sns.lineplot(data=iris.drop(['species'],axis=1))
```

Draw bar plot with Matplotlib and Seaborn

```python
fig ,ax1 =plt.subplots()

data=iris['petal_length'].value_counts()

points=data.index
frequency=data.values

 #create a bar chart
ax1.bar(points, frequency)

 #set title and labels
ax1.set_title('Petal Length')
ax1.set_xlabel('Petal length')
ax1.set_ylabel("Frequency")
```

```python
sns.countplot(iris['petal_length'])
```

Scatter plot with matplotlib and seaborn 

```python
fig,ax2=plt.subplots()

#scatter the sepal length against the sepal _width

ax2.scatter(iris['sepal_length'],iris['petal_width'])

ax2.set_title('Iris Dataset')
ax2.set_xlabel('sepal_length')
ax2.set_ylabel('sepal_width')
```

```python
colors={'setosa':'r','versicolor':'g','virginica':'b'}
fig,ax3=plt.subplots()

#scatter the sepal length against the sepal _width
for i in range(len(iris['sepal_length'])):

  ax3.scatter(iris['sepal_length'][i],iris['sepal_width'][i],color=colors[iris['species'][i]])

ax3.set_title('Iris Dataset')
ax3.set_xlabel('sepal_length')
ax3.set_ylabel('sepal_width')
```
