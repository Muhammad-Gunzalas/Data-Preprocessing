### Creating a logistic regression to predict absenteeism 

import the relevant libraries

```python
import pandas as pd
import numpy as np
```

```python
data_preprocessed=pd.read_csv('Absenteeism_preprocessed.csv')
data_preprocessed
```

```python
data_preprocessed.head()
```

### create a target

```python
data_preprocessed['Absenteeism Time in Hours'].median()
```

```python
targets=np.where(data_preprocessed['Absenteeism Time in Hours']>
                 data_preprocessed['Absenteeism Time in Hours'].median(),1,0)
targets
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
```

```python
data_preprocessed.head()
```

### A Common on the targets

```python
targets.sum()/targets.shape[0]
```

```python
data_with_targets  = data_preprocessed.drop(['Absenteeism Time in Hours'],axis=1)
```

```python
data_with_targets is data_preprocessed
```

```python
data_with_targets.head()
```

```python
data_with_targets.shape
```

### select inputs for the regression

```python
data_with_targets.iloc[:,0:14]
```

```python
data_with_targets.iloc[:,:-1]
```

```python
uncaled_inputs=data_with_targets.iloc[:,:-1]
```

### standerdize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
absenteeeism_scaler = StandardScaler()
```

```python
absenteeeism_scaler.fit(uncaled_inputs)
```

```python
scaled_inputs = absenteeeism_scaler.transform(uncaled_inputs)
```

```python
scaled_inputs
```

```python
scaled_inputs.shape
```

### split the data into train & test and shuffle

### import the relevant model

```python
from sklearn.model_selection import train_test_split
```

### split

```python
train_test_split(scaled_inputs,targets)
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_inputs,targets,train_size=0.8)
```

```python
print(x_train.shape,y_train.shape)
```

```python
print(x_test.shape,y_test.shape)
```

### Logistic Regression with sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

### training model

```python
reg=LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

### manually check the accuracy

```python
model_outputs=reg.predict(x_train)
```
